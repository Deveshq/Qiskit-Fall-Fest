```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#convert the image into plot of data points

image = cv2.imread('images.png', cv2.IMREAD_GRAYSCALE)
ret, image = cv2.threshold(image, 128, 128, cv2.THRESH_BINARY)

scatter = np.where(image==0)
x,y = scatter[1],scatter[0]
display(x,y)

plt.scatter(x,y,color='blue',marker='.',s=0.2)  
plt.show()
```
